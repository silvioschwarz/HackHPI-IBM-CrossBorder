```python
import plotly.plotly as py
import plotly.graph_objs as go 

data = pd.read_csv('data/TerrorAfricaTime.csv')
trace = go.Scattergl(x=data.event_date,
                   y=data.fatalities)

data = [trace]
layout = dict(
    title='Time series with range slider and selectors',
    yaxis=dict(
        range=[0, 2000],
        autorange=False,
        showgrid=True,
        zeroline=True,
        showline=True,
        ticks='',
        showticklabels=True
    ),
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(count=1,
                    label='YTD',
                    step='year',
                    stepmode='todate'),
                dict(count=1,
                    label='1y',
                    step='year',
                    stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

from IPython.display import IFrame
fig = dict(data=data, layout=layout)
#display(IFrame(fig, 20, 20))
py.iplot(fig)
```
